```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def detect_curves(image_path):
    # Read the image
    image = cv2.imread(image_path)

    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur to reduce noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Perform edge detection
    edges = cv2.Canny(blurred, 50, 150)

    # Find contours in the edge-detected image
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw detected contours on the original image
    cv2.drawContours(image, contours, -1, (0, 255, 0), 3)

    # Display the resulting image using Matplotlib
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```

```python
def main():
    # Path to the image
    image_path = '/content/curve image.png'  # Replace with the path to your image file

    # Detect curves in the image
    detect_curves(image_path)

if __name__ == "__main__":
    main()
```
